```python
import pandas as pd
from sklearn.datasets import make_classification
import numpy as np
import math
from sklearn.metrics import adjusted_rand_score,rand_score
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
```

```python
iris=load_iris()

x=iris.data[:,:2]
y=iris.target

plt.figure()
plt.scatter(x[:,0],x[:,1])
```

```python
from sklearn.datasets import make_blobs

centers = [[1, 1], [-1, -1], [1, -1]]
x, y = make_blobs(n_samples=1000, centers=centers, cluster_std=0.5, random_state=123)

plt.scatter(x[:, 0], x[:, 1])
plt.show()
```

## K-means Clustering

```python
from sklearn.cluster import KMeans

kmmodel=KMeans(n_clusters=3,max_iter=500)

y_pred=kmmodel.fit_predict(x)

#print(x)

#print(y)
#print(y_pred)

print(rand_score(y,y_pred))
print(adjusted_rand_score(y,y_pred))

res=np.unique(y_pred)

print(res)

plt.figure()

for clust in res:
    row=np.where(y_pred==clust)
    plt.scatter(x[row,0],x[row,1])
```

## DBSCAN

```python
from sklearn.cluster import DBSCAN

dbsmodel=DBSCAN(min_samples=13,eps=0.2)

y_pred=dbsmodel.fit_predict(x)

#print(x)

#print(y)
#print(y_pred)

print(rand_score(y,y_pred))
print(adjusted_rand_score(y,y_pred))

res=np.unique(y_pred)

print(res)

plt.figure()

for clust in res:
    row=np.where(y_pred==clust)
    plt.scatter(x[row,0],x[row,1])
```

## HDBSCAN

```python
from sklearn.cluster import HDBSCAN

hdbsmodel=HDBSCAN()

y_pred=hdbsmodel.fit_predict(x)

#print(x)

#print(y)
#print(y_pred)

print(rand_score(y,y_pred))
print(adjusted_rand_score(y,y_pred))

res=np.unique(y_pred)

print(res)

plt.figure()

for clust in res:
    row=np.where(y_pred==clust)
    plt.scatter(x[row,0],x[row,1])
```

## Afinity Propagation

```python
from sklearn.cluster import AffinityPropagation

apmodel=AffinityPropagation(damping=.5,max_iter=1000,convergence_iter=100)

y_pred=apmodel.fit_predict(x)

#print(x)

#print(y)
#print(y_pred)

print(rand_score(y,y_pred))
print(adjusted_rand_score(y,y_pred))

res=np.unique(y_pred)

print(res)

plt.figure()

for clust in res:
    row=np.where(y_pred==clust)
    plt.scatter(x[row,0],x[row,1])
```
