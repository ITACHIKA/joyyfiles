```python
import pandas as pd
from sklearn.model_selection import train_test_split
import numpy as np
import math
from sklearn.datasets import load_diabetes
from sklearn.metrics import r2_score
import matplotlib.pyplot as plt
```

```python
# diabete dataset
dbtData=load_diabetes()

x=dbtData.data
print(x)
y=dbtData.target

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=91)
```

```python
#california housing dataset
from sklearn.datasets import fetch_california_housing


chData=fetch_california_housing()

x=chData.data
print(x)
y=chData.target

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=123213)
```

## Linear Regression

```python
from sklearn.linear_model import LinearRegression

linearModel=LinearRegression()

linearModel.fit(x_train,y_train)

y_pred=linearModel.predict(x_test)

print(y_test)
print(y_pred)

print(r2_score(y_test,y_pred))

a=plt.figure(figsize=[10.24,7.68])
print(a.dpi)
plt.scatter(y_pred,y_test,color='red')
plt.xlabel("pred")
plt.ylabel("true")
plt.plot([0,5],[0,5])

plt.show()
```

## SGD Regression

```python
from sklearn.linear_model import SGDRegressor

sgdModel=SGDRegressor(alpha=0.0001,max_iter=100000)

sgdModel.fit(x_train,y_train)

y_pred=sgdModel.predict(x_test)

print(r2_score(y_test,y_pred))


plt.figure()
plt.scatter(y_pred,y_test,color='red')
plt.xlabel("pred")
plt.ylabel("true")
plt.plot([0,5],[0,5])

plt.show()
```

## Lasso Regression


```python
from sklearn.linear_model import Lasso

lassoModel=Lasso(max_iter=1000,alpha=0.01)

lassoModel.fit(x_train,y_train)

y_pred=lassoModel.predict(x_test)

print(r2_score(y_test,y_pred))

print(lassoModel.coef_)

plt.figure()
plt.scatter(y_pred,y_test,color='red')
plt.xlabel("pred")
plt.ylabel("true")
plt.plot([0,5],[0,5])

plt.show()
```

## Ridge Regression

```python
from sklearn.linear_model import Ridge

ridgeModel=Ridge(max_iter=10000,alpha=0.01)

ridgeModel.fit(x_train,y_train)

y_pred=ridgeModel.predict(x_test)

print(r2_score(y_test,y_pred))

print(ridgeModel.coef_)

plt.figure()
plt.scatter(y_pred,y_test,color='red')
plt.xlabel("pred")
plt.ylabel("true")
plt.plot([0,6],[0,6])

plt.show()
```

## Decision Tree regression

```python
from sklearn.tree import DecisionTreeRegressor

dtModel=DecisionTreeRegressor(max_depth=10)

dtModel.fit(x_train,y_train)

y_pred=dtModel.predict(x_test)

print(r2_score(y_test,y_pred))


plt.figure()
plt.scatter(y_pred,y_test,color='red')
plt.xlabel("pred")
plt.ylabel("true")
plt.plot([0,5],[0,5])

plt.show()
```

## K Nearest Neighbors Regression

```python
from sklearn.neighbors import KNeighborsRegressor

knnrModel=KNeighborsRegressor(n_neighbors=5)

knnrModel.fit(x_train,y_train)

y_pred=knnrModel.predict(x_test)

print(r2_score(y_test,y_pred))


plt.figure()
plt.scatter(y_pred,y_test,color='red')
plt.xlabel("pred")
plt.ylabel("true")
plt.plot([0,5],[0,5])

plt.show()
```

## Radius Neighbour regressor

```python
from sklearn.neighbors import RadiusNeighborsRegressor

rnnModel=RadiusNeighborsRegressor(radius=6)

rnnModel.fit(x_train,y_train)

y_pred=rnnModel.predict(x_test)

print(len(y_pred))
del_list=[]
for i in range(len(y_pred)-1,0,-1):
    if np.isnan(y_pred[i]):
        del_list.append(i)
print(del_list)
print(np.shape(y_pred))
y_test=np.delete(y_test,del_list,axis=0)
y_pred=np.delete(y_pred,del_list,axis=0)
print(y_test)
print(y_pred)

print(r2_score(y_test,y_pred))


plt.figure()
plt.scatter(y_pred,y_test,color='red')
plt.xlabel("pred")
plt.ylabel("true")
plt.plot([0,5],[0,5])

plt.show()
```

## Randrom Forest Regression

```python
from sklearn.ensemble import RandomForestRegressor

rfrModel=RandomForestRegressor(n_estimators=50)

rfrModel.fit(x_train,y_train)

y_pred=rfrModel.predict(x_test)

print(r2_score(y_test,y_pred))


plt.figure()
plt.scatter(y_pred,y_test,color='red')
plt.xlabel("pred")
plt.ylabel("true")
plt.plot([0,5],[0,5])

plt.show()
```

## SVM regression

```python
from sklearn.svm import SVR

svrModel=SVR()

svrModel.fit(x_train,y_train)

y_pred=svrModel.predict(x_test)

print(r2_score(y_test,y_pred))


plt.figure()
plt.scatter(y_pred,y_test,color='red')
plt.xlabel("pred")
plt.ylabel("true")
plt.plot([0,5],[0,5])

plt.show()
```
